```python
import pandas as pd
import geopandas as gpd
```

```python
cantones = gpd.read_file("datos/cantones-crtm05.geojson")
cantones
```

```python
cantones.plot()
```

```python
rvn = gpd.read_file("datos/redvial-crtm05.geojson")
rvn
```

```python
rvn.plot()
```

```python
interseccion_redvial_cantones = gpd.overlay(rvn, cantones, how='intersection')
interseccion_redvial_cantones
```

```python
interseccion_redvial_cantones["longitud_rvn"] = interseccion_redvial_cantones["geometry"].length/1000
interseccion_redvial_cantones[["categoria", "codigo", "canton", "geometry", "longitud_rvn"]]
```

```python
# Agrupación y suma
suma_redvial_cantones = interseccion_redvial_cantones.groupby(["cod_canton"])["longitud_rvn"].sum()

# Para convertir la serie en dataframe
suma_redvial_cantones = suma_redvial_cantones.reset_index()

suma_redvial_cantones
```

## Calculo de densidad red vial

```python
# Se le agrega a la capa de cantones la longitud de la red vial para cada polígono
cantones_redvial = cantones.merge(suma_redvial_cantones, on="cod_canton")

# Se calcula el área del cantón
cantones_redvial["area_canton"] = cantones_redvial["geometry"].area/1000000

# Se calcula la densidad
cantones_redvial["densidad_rvn"] = cantones_redvial["longitud_rvn"] / cantones_redvial["area_canton"]

cantones_redvial.sort_values(by="densidad_rvn", ascending=False)
```

```python
# Se le agrega a la capa de cantones la longitud de la red vial para cada polígono
cantones_redvial = cantones.merge(suma_redvial_cantones, on="cod_canton")

# Se calcula el área del cantón
cantones_redvial["area_canton"] = cantones_redvial["geometry"].area/1000000

# Se calcula la densidad
cantones_redvial["densidad_redvial"] = cantones_redvial["longitud_rvn"] / cantones_redvial["area_canton"]

cantones_redvial.sort_values(by="densidad_redvial", ascending=False)
```

## Salidas

```python
cantones_redvial.plot(column="densidad_redvial",
                      legend=True,
                      cmap='Reds', 
                      edgecolor="black",
                      scheme='quantiles',
                      figsize=(20, 20))
```
